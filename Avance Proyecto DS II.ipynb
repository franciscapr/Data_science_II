#### Diccionario de Datos:

- **Suburb**: Nombre del suburbio donde se encuentra la propiedad.
- **Rooms**: Número de habitaciones en la propiedad.
- **Type**: Tipo de propiedad, con las siguientes categorías:
  - `br`: Dormitorio(s)
  - `h`: Casa, cabaña, villa, semi, terraza
  - `u`: Unidad, dúplex
  - `t`: Casa adosada
  - `dev site`: Sitio de desarrollo
  - `o res`: Otro residencial
- **Price**: Precio de venta en dólares.
- **Method**: Método de venta de la propiedad:
  - `S`: Propiedad vendida
  - `SP`: Propiedad vendida anteriormente
  - `PI`: Propiedad transmitida
  - `PN`: Vendido antes no divulgado
  - `SN`: Vendido no divulgado
  - `NB`: Ninguna oferta
  - `VB`: Oferta del proveedor
  - `W`: Retirado antes de la subasta
  - `SA`: Vendido después de la subasta
  - `SS`: Precio vendido después de la subasta no revelado
  - `N/A`: Precio o oferta más alta no disponible
- **SellerG**: Agente de bienes raíces.
- **Date**: Fecha de la venta.
- **Distance**: Distancia desde el Distrito Central de Negocios (CBD). Centro de la ciudad.
- **Postcode**: Código postal de la propiedad.
- **Bedroom2**: Número de dormitorios extraído de una fuente diferente.
- **Bathroom**: Número de baños en la propiedad.
- **Car**: Número de plazas de aparcamiento.
- **Landsize**: Tamaño del terreno en metros cuadrados.
- **BuildingArea**: Tamaño del edificio en metros cuadrados.
- **YearBuilt**: Año de construcción de la propiedad.
- **CouncilArea**: Consejo rector del área.
- **Lattitude**: Latitud geográfica de la propiedad.
- **Longtitude**: Longitud geográfica de la propiedad.
- **Regionname**: Región general (Oeste, Noroeste, Norte, Noreste, etc.).
- **Propertycount**: Número de propiedades que existen en el suburbio.

# Eliminación Variable

```python
# Borraremos la variable que no se utilizara
df = df.drop('N_dormitorios_f_externa', axis=1)
```

# Imputación por la Mediana

```python
# Imputaremos las variables numericas que contiene menos valores nulos
```

```python
col_inp = df[['CBD', 'Codigo_postal', 'N_propiedades_suburbio']]
```

```python
col_inp.head(3)
```

```python
for i in col_inp:    # Reemplazar por la mediana en las columnas numericas de col_interpolacion
    df[i] = df[i].fillna(df[i].median())
```

```python
# Verificamos los valores nulos
df.isnull().sum()
```

# Imputación variables categoricas

```python
col_inp_categorica = df[['Consejo_rector', 'Region_general']]
```

```python
for i in col_inp_categorica:
    df[i] = df[i].fillna(df[i].mode().iloc[0])
```

```python
df.isnull().sum()
```

# Imputación varibles numericas

```python
# Asignamos las columnas que queremos imputar 
numeric_columns = ['T_edificio_m2', 'Ano_construccion', 'Tamano_terreno_m2', 'Latitud_geografica',
                        'Longitud_geografica', 'Precio', 'N_banos', 'N_aparcamiento']
# Calculamos la mediana de las columna agrupadas por suburbio y las asignamos a una variable medianas
grouped_median = df.groupby('CBD')[numeric_columns].transform('median')
# convertimos la variable en un diccionar
df[numeric_columns] = df[numeric_columns].fillna(grouped_median)

grouped_median.head(3)
```

## Frecuencia de Propiedades por Región

```python
# Creamos un histograma con seaborn
sns.histplot(df['Region_general'])

# Rotamos las etiquetas del eje x
plt.xticks(rotation=45, ha='right')
plt.xlabel('Región General')
plt.ylabel('Frecuencia de Propiedades')
plt.title('Histograma de Región General')
plt.tight_layout()    # Ajustamos el diseño para que todo se vea bién

plt.grid(True)
plt.show()
```

- En Southern Metropolitan y Northern Metropolitan, se encuentran las mayores concentraciones de propiedades de las distintas regiones de Australia.

## Precio promedio de Propiedad según Región

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='Region_general', y='Precio', data=df, estimator=np.mean)

# Añadir títulos y etiquetas
plt.title('Precio Promedio por Región')
plt.xlabel('Región')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=45, ha='right')

plt.grid(True)
plt.show()
```

- La regiones Southern Metropolitan y Easter Metropolitan tiene precios promedio mayores a los de las demás regiones.
- Western Victoria y Northerm Victoria tiene precios promedio menores al resto de las demas regiones.

## Precio propiedad según proximidad al CBD

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(data=df, x='CBD', y='Precio', alpha=0.5)

# Añadir títulos y etiquetas
plt.title('Impacto de la Proximidad al CBD en el Precio de la Propiedad')
plt.xlabel('Distancia al CBD (km)')
plt.ylabel('Precio de la Propiedad')

plt.grid(True)
plt.show()
```

- A menor distancia del centro CBD es mayor el precio de la propiedad, es decir en un rango de entre 0 a 11 kilometros de distancia los valores de las propiedades son elevados, por el contrario los precios comienzan a descender significativamente desde los 12 km hasta los 50 km de distancia desde el CBD. 

## Cantidad de Compras por Año

```python
# Creamos una nueva variable año para asignarle el año que fueron vendidas las propiedades
df['año'] = df['Fecha_venta'].dt.year
```

```python
conteo_por_año = df.groupby('año').size().reset_index(name='cantidad')
print(conteo_por_año)

sns.barplot(x='año', y='cantidad', data=conteo_por_año)
plt.xlabel('Año')
plt.ylabel('Cantidad de Registros')
plt.title('Cantidad de Ventas por Año')
plt.show()
```

- Podemos observar que el año 2017 presenta muchas más ventas de propiedades que los demas año.
